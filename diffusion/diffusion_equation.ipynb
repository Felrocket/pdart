Diffusion Equation from Bastin et al (1971, Equation 2) 
\begin{equation}
I = \frac{A}{(kt)^{\frac{3}{2}}}\exp\left[ -\frac{r^2}{4kt} - \frac{2\pi ft}{Q} \right]
\end{equation}

```python
# Constants from Latham 1970 (also used in Bastin 1971)
Q = 3600
k = 1.2e6 # m^2/s
# mean free path on the order of 300 m. 
A = 1

r = 100000 #m 
f = 0.45 # Hz

def diffusion_I_firstterm(t,A,k,r,f,Q):
    
    I = A/((k*t)**(3/2)) * np.exp(-((r**2)/(4*k*t)))
    return(I)
    
def diffusion_I(t,A,k,r,f,Q,normalized=False):
    
    I = A/((k*t)**(3/2)) * np.exp(-((r**2)/(4*k*t)) - ((2*np.pi*f*t)/Q))
    if normalized:
        max_I = np.nanmax(I)
        I = I/max_I
    return(I)

    
def diffusion_I_first_term(t,A,k,r,f,Q,normalized=False):
    
    I = np.exp(-((r**2)/(4*k*t)))
    if normalized:
        max_I = np.nanmax(I)
        I = I/max_I
    return(I)

def diffusion_I_second_term(t,A,k,r,f,Q,normalized=False):
    
    I = np.exp(-(2*np.pi*f*t)/Q)
    if normalized:
        max_I = np.nanmax(I)
        I = I/max_I
    return(I)

def diffusion_I_exponent(t,A,k,r,f,Q,normalized=False):
    
    I = np.exp(-((r**2)/(4*k*t)) - ((2*np.pi*f*t)/Q))
    if normalized:
        max_I = np.nanmax(I)
        I = I/max_I
    return(I)

def diffusion_I_outer(t,A,k,r,f,Q,normalized=False):
    
    I = A/((k*t)**(3/2))
    if normalized:
        max_I = np.nanmax(I)
        I = I/max_I
    return(I)
```

```python
def calc_t_max(A,k,f,Q,start_t0=False):
    # time array, in seconds     
    t = np.arange(0.5,3600*4)
    # distance array, in km    
    dist = np.arange(5,1005,5)
    # array for time to t_max     
    t_max_array = np.zeros(len(dist))

    for i, r in enumerate(dist): 
        # distance r in meters          
        r = r*1000
        # calculate the amplitude of the diffusion equation         
        I = A/((k*t)**(3/2)) * np.exp(-((r**2)/(4*k*t)) - ((2*np.pi*f*t)/Q))
        # find the maximum (note this only returns the closest index, 
        # so requires a lot of points in the time array)        
        max_index = numpy.where(I == numpy.amax(I))
        # the time to tmax for each distance 
        t_max_array[i] = t[max_index]
    
    return(dist, t_max_array)

def calc_t_onset_t_max(A,k,f,Q,start_t0=False):
    # time array, in seconds     
    t = np.arange(0.5,3600*4)
    # distance array, in km    
    dist = np.arange(5,1005,5)
    # array for time to t_max     
    t_max_array = np.zeros(len(dist))

    for i, r in enumerate(dist): 
        # distance r in meters          
        r = r*1000
        # calculate the amplitude of the diffusion equation         
        I = A/((k*t)**(3/2)) * np.exp(-((r**2)/(4*k*t)) - ((2*np.pi*f*t)/Q))
        # normalize
        max_I = np.nanmax(I)
        I = I/max_I
        
        # find the start - this is a bit subjective, but try as a percentage of the max
#       #(note this only returns the closest index, 
        # so requires a lot of points in the time array) 
        start_index = np.argmax(I>0.01)
    
        # find the maximum (note this only returns the closest index, 
        # so requires a lot of points in the time array)        
        max_index = numpy.where(I == numpy.amax(I))

        # the time to tmax for each distance 
        t_max_array[i] = t[max_index] - t[start_index]
    
    return(dist, t_max_array)

# plot epicentral distance versus time to tmax (from the origin time t0) to 
fig = plt.figure()
gs = gridspec.GridSpec(2,2)
# fig, ax = plt.subplots(figsize=(8, 4))

freqs = [0.5, 1, 1.5]

ax1 = fig.add_subplot(gs[0])

colors = plt.cm.winter(np.linspace(0,1,len(freqs)))

for i, f in enumerate(freqs): 
    dist, t_max_array = calc_t_max(A=1,k=1.2e6,f=f,Q=3600,
        start_t0=True)
#     plot the time to t_max in minutes?
    ax1.plot(dist,t_max_array/60,color=colors[i],linewidth=1.3,label=str(f))

ax1.set_xlabel('Distance [km]')
ax1.set_ylabel('[minutes]')
ax1.set_xlim(0,350)
ax1.set_ylim(0, 110)
ax1.legend(ncol=3,prop={'size': 10},title=r'f [Hz]', framealpha=0,
           columnspacing=0.2,loc='upper left')
ax1.set_title('t_0 to t_max')

ax2 = fig.add_subplot(gs[1])

colors = plt.cm.winter(np.linspace(0,1,len(freqs)))

for i, f in enumerate(freqs): 
    dist, t_max_array = calc_t_onset_t_max(A=1,k=1.2e6,f=f,Q=3600,
        start_t0=True)
#     plot the time to t_max in minutes?
    ax2.plot(dist,t_max_array/60,color=colors[i],linewidth=1.3,label=str(f))

ax2.set_xlabel('Distance [km]')
ax2.set_ylabel('[minutes]')
ax2.set_xlim(0,350)
ax2.set_ylim(0, 60)
ax2.legend(ncol=3,prop={'size': 10},title=r'f [Hz]', framealpha=0,
           columnspacing=0.2,loc='upper left')
ax2.set_title('t_onset to t_max')


# plt.suptitle('Time to t_max', size=16)
gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
plt.show()


print('''The time from t_onset to t_max is something we can measure. 
This code varies plots the time against distance for each of the frequencies. 
Note that the other parameters are being held constant, which may not reflect reality.''')
```
